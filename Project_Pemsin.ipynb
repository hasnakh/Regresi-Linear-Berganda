```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('Data Penjualan Ayam.csv')
data.head()
```

```python
data.info()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
data['year']=le.fit_transform(data['tahun'])
data.head(10)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data['jumlah_ayam']=scaler.fit_transform(data[['jumlah_ayam']])
data['harga_unit']=scaler.fit_transform(data[['harga_unit']])
data['total_harga']=scaler.fit_transform(data[['total_harga']])
data.head(10)
```

```python
databaru=data[['year','jumlah_ayam','harga_unit','total_harga']]
databaru.head()
```

```python
x = databaru[['year','harga_unit','total_harga']]
y = databaru[['jumlah_ayam']].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=4)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
```

```python
regr.fit(X_train, y_train)
```

```python
regr.predict(X_test)
```

```python
ypred = regr.predict(X_test)
ypred
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

print("r1 score : ", r2_score(y_test, ypred))
print("MAE score: ", mean_absolute_error(y_test, ypred))
print("MSE score: ", mean_squared_error(y_test, ypred))
```

```python
print ('Coefficients (b) : ', regr.coef_)
print ('Intercept (a) : ', regr.intercept_)
```
